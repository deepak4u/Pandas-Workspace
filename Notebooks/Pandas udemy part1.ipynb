### 2. Get all rows that contain one substring OR another substring

```python
bond[bond["Actor"].str.contains("Sean Connery|Roger Moore")]
```

```python
#  apply the â€˜~â€™ symbol before the df[â€˜Actorâ€™]:
```

```python
bond[~bond["Actor"].str.contains("Sean Connery|Roger Moore|Pierce Brosnan")]
```

## Outer join

```python
# Remove duplicate values - less value than concat 2 table, because it removes same values
week1.merge(week2, how='outer', on='Customer ID', suffixes=[' - Week 1', ' - Week 2'])
```
